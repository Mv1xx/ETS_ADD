Crie um violinplot de tip por day, diferencie o gênero pela cor. Responda: Em quais dias há maior variação nas gorjetas?

```python
tips = pd.read_csv("../../CSV/tips.csv", sep=';')

sns.violinplot(x='day', y='tip', hue='sex',data=tips, split=True)
```
