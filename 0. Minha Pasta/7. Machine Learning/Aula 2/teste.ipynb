```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
casas = pd.read_csv("house_prices.csv")
```

```python
casas.drop('date', axis = 1, inplace = True) #dificil basear outras caracteristicas na data
#casas.describe() ->> caracteristicas do dataset
casas.isnull().sum() #qnt de valores nulos em cada coluna
```

```python
x_casas = casas.iloc[:, 2:18]
y_casas = casas.iloc[:, 1].values
```

```python
from sklearn.model_selection import train_test_split ##separa o dataset em treino e teste
x_casa_treino, x_casa_teste, y_casa_treino, y_casa_teste = train_test_split(x_casas, y_casas, test_size = 0.2, random_state=42) #random_state >> aleatoriedade

```

```python
regressor_multiplo_casa = LinearRegression()
regressor_multiplo_casa.fit(x_casa_treino, y_casa_treino) #faz o treinamento

regressor_multiplo_casa.score(x_casa_treino, y_casa_treino) #resultado de assertividade
regressor_multiplo_casa.score(x_casa_teste, y_casa_teste) #resultado de assertividade
```

Overfitting >> bem no treinamento e mal no teste >> preciso somente em valores que ele conhece
Underfitting >> mal no treinamento e no teste >> dados distantes da correlação

```python
previsoes = regressor_multiplo_casa.predict(x_casa_teste)
previsoes
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import math

mae = mean_absolute_error(y_casa_teste, previsoes)
print("MAE ", mae)

mse = mean_squared_error(y_casa_teste, previsoes)
print("MSE ", mse)

rmse = math.sqrt(mean_squared_error(y_casa_teste, previsoes))
print("RMSE ", rmse)
```
