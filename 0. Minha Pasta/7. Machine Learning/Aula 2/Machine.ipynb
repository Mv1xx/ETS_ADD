# Previsão Precisão Linear

P = b + m * v

Legenda:
 
 P = N° a ser previsto (y)

 b = Intersecção, o valor de y que a reta corta o eixo y

 m = inclinação
 
 v = valor de x a ser previsto

# Calcular CV:
cv = somatorio((xi - mediax) * (yi - mediay)) / numero de elementos

# Calcular CR:
cr = cv / (Std(y) * Std(x))

# Calcular m:
m = CR * (Std(y) / Std(x))

# Calcular b:
b = média y - média x * m

```python
import pandas as pd
import math as m
import numpy as np

plano = pd.read_csv("plano_saude.csv", sep=",")

media_idades = plano['idade'].mean()
media_custos = plano['custo'].mean()

def calcular_cv():
    somatorio = 0

    for i in range(plano.count()['custo']):
        custo = plano.iloc[i]['custo']
        idade = plano.iloc[i]['idade']

        somatorio += (custo - media_custos) * (idade - media_idades)
    
    cv = np.corrcoef(plano['custo'], plano['idade'])

    return somatorio / (plano.count()['custo'] - 1)

def calcular_cr():
    # Sem numpy

    # somatorio_custo = 0
    # somatorio_idade = 0

    # for i in range(plano.count()['custo']):
    #     custo = plano.iloc[i]['custo']
    #     idade = plano.iloc[i]['idade']

    #     somatorio_custo += m.pow(float(custo) - float(media_custos), 2)
    #     somatorio_idade += m.pow(idade - media_idades, 2)

    # desviop_custo = m.sqrt(somatorio_custo / plano.count()['custo'])
    # desviop_idade = m.sqrt(somatorio_idade / plano.count()['custo'])

    desviop_custo = np.std(plano['custo'])
    desviop_idade = np.std(plano['idade'])

    return calcular_cv() / (desviop_custo * desviop_idade)

def calcular_m():
    # somatorio_custo = 0
    # somatorio_idade = 0

    # for i in range(plano.count()['custo']):
    #     custo = plano.iloc[i]['custo']
    #     idade = plano.iloc[i]['idade']

    #     somatorio_custo += m.pow(float(custo) - float(media_custos), 2)
    #     somatorio_idade += m.pow(idade - media_idades, 2)

    # desviop_custo = m.sqrt(somatorio_custo / plano.count()['custo'])
    # desviop_idade = m.sqrt(somatorio_idade / plano.count()['custo'])

    desviop_custo = np.std(plano['custo'])
    desviop_idade = np.std(plano['idade'])

    return calcular_cr() * (desviop_custo / desviop_idade)

def calcular_b():
    return media_custos - media_idades * calcular_m()

def calcular_p(valor):
    return calcular_b() + calcular_m() * valor
```

```python
print(calcular_p(18))
print(calcular_p(23))
print(calcular_p(67))
```
