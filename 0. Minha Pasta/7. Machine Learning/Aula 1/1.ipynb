```python
import pandas

plano_saude = pandas.read_csv("../Aula 1/plano_saude.csv")


def covariancia():
    mediax = round(plano_saude["idade"].mean(), 2)
    mediay = round(plano_saude["custo"].mean(), 2)
    soma = 0
    
    for i in range(len(plano_saude.index)):
        x = plano_saude["idade"][i] - mediax
        y = plano_saude["custo"][i] - mediay
        
        soma += (x * y) 
    
    cv = soma / plano_saude.index - 1
    return cv

def coeficiente_variacao():
    cv = covariancia()
    cr = cv / (plano_saude["idade"].std() * plano_saude["custo"].std())
    
    return cr

resultado = covariancia()
print(resultado)
    
    


```
